**3. Descripción de los datos**

Cada fila representa una reserva de hotel urbano o resort efectada correctamente o cancelada y cada columna representa variables que definen el estado y los detalles de la reserva de la siguiente manera:

*  **hotel:** Variable Categorica con dos valores unicos que representa "Resort hotel" para el hotel resort y "City hotel" para el hotel urbano.

*  **is_canceled:** Variable Categorica con dos valores unicos que represente 1 cuando la reserva fue cancelada y 0 cuando la reserva no fue cancelada.

*   **lead_time**: Variable númerica que representa el número de días transcurridos entre la fecha de entrada de la reserva en el base de datos del hotel y la fecha de llegada al hotel.

*   **arrival_date_year:** Variable númerica que representa el año de llegada al hotel.

* **arrival_date_month:** Variable Categorica que presenta el mes de llegada al hotel.

* **arrival_date_week_number:** Variable númerica que representa el número de semana de llegada al hotel.

* **arrival_date_day_of_month:** Variable númerica que representa un día en el mes especifico de llegada al hotel.

* **stays_in_weekend_nights:** Variable númerica que representa el número de noches de fin de semana (sábado o domingo) que el huésped se alojó o reservó para quedarse en el hotel.

* **stays_in_week_nights:** Variable númerica que represente el número de noches de la semana (de lunes a viernes) que el huésped se hospedó o reservó para quedarse en el hotel.

* **adults:** Variable númerica que representa la cantidad de adultos que se hospedaran en el hotel de acuerdo a la reserva del hotel.  

* **children:** Variable númerica que representa la cantidad de niños que se hospedaran en el hotel de acuerdo a la reserva del hotel.

* **babies:** Variable númerica que representa la cantidad de bebés que se hospedaran en el hotel de acuerdo a la reserva del hotel.

* **meal:** Variable categorica que represente el tipo de comida reservada. Las categorías se presentan en paquetes de comida: Sin definir/SC, sin paquete de comida; AD, Alojamiento y Desayuno; MP – Media pensión (desayuno y otra comida – normalmente cena); FB, Pensión completa (desayuno, comida y cena).

* **country:** Variable categorica que representa el país desde el cual se realiza la reserva, tiene formato ISO 3155–3:2013.

* **market_segment:** Variable categorica que representa la designación del segmento de mercado. En las categorías, el término "TA" significa "Agentes de viajes" y "TO" significa "Operadores turísticos".

* **distribution_channel:** Variable categorica que representa el canal de distribución de reservas. El término "TA" significa "Agentes de viajes" y "TO" significa "Operadores turísticos".

* **is_repeated_guest:** Variable Categorica que representa si la reserva fue realizada por un cliente repetido o que ya hubiera visitado los hoteles anteriormente. La varible mas 1 cuando es un cliente repretido y 0 cuando el cliente es nuevo.

* **previous_cancellations:** Vairbale númerica que presenta el número de reservas anteriores canceladas por parte de un invitado que haya visitado los hoteles mas de una vez.  

* **previous_bookings_not_canceled:** Variable númerica que presenta el número de reservas anteriores no canceladas por parte de un invitado que haya visitado los hoteles mas de una vez.

* **reserved_room_type:** Variable categorica que presenta un codigo con el tipo de habitación reservada.

* **assigned_room_type:** Variable categorica que presenta un codigo con el tipo de habitación asignada, este variable puede variar del tipo de habitiación reservada por razones de funcionamiento del hotel.

* **booking_changes:** Variable Númerica que presenta el número de modificaciones a la reserva que entra a la base de datos hasta el momento del check-in o cancelación. (Este incluye cantidad de personas, tipo de habitación, comidas, fechas de llegada, cantidad de noches).

* **deposit_type:** Variable Categorica que indicado si el cliente realizo un deposito para garantizar la reserva o no. se encuentran tes categorias No Deposit, indica que el cliente no realizo ningun deposito, Non Refund, indica que se realiza un deposito del valor total de la estadia; Refundable, indica que realiza un reembolso por el valor total del costo de la estadia.

* **agent:** Variable categorica que indica el ID de la agencia de viajes que realizo la reserva.

* **company:** Variable categorica que indica el ID de la empresa o enetidad que realizo la reserva o responsable del pago de la reserva.

* **days_in_waiting_list:** Variable númerica que indica el número de días que la reserva estuvo en lista de espera antes de ser confirmada.

* **customer_type:** Variable categorica que indicado que el tipo de cliente que realiza la reserva, se clasifican en cuantro, el primero es Contract, cuando la reserva tiene asociada una asignación u otro tipo de contrato; Segundo es Group, cuando la reserva esta asociada a un grupo; Tercero es Transient, cuando la reserva no es parte de un grupo o contrato, y no está asociada a otra reserva transitoria; y la cuarta Transient-party, cuando la reserva es transitoria, pero está asociada al menos a otra reserva transitoria.  

* **adr:** Variable Númerica que indica la tarifa diaria promedio, al dividir la suma de todas las transacciones de alojamiento con el número total de noches de estadia.

* **required_car_parking_spaces:** Variable númerica que indicado el número de paruqeaderos requeridos por el cliente.

* **total_of_special_requests:** Variable númerica que indica el número de solicitudes especiales realizadas por el cliente como un cama doble o un piso alto.

* **reservation_status:** Variable Categorica que indica el ultimo estado de la reserva, Caceled, si esta fue cancelada por el cliente; Check-Out, el cliente se registro pero ya partio; No-Show, el cliente no se registro pero informo el motivo. 

* **reservation_status_date:** Variable númerica que indica le fecha en la que se establecio el ultimo estado de la reserva.

Se ve la necesidad de aplciar hot_encodign en la variable hotal ya que tiene dos categorias que no tienen un orden especifico

```python

# importing one hot encoder 
from sklearn.preprocessing import OneHotEncoder

# Creamos el objeto
onehotencoder = OneHotEncoder()

#reshape the 1-D country array to 2-D as fit_transform expects 2-D and finally fit the object 

X = onehotencoder.fit_transform(df.hotel.values.reshape(-1,1)).toarray()

X
```

```python
dfOneHot = pd.DataFrame(X, columns = ["Hotel_"+str(int(i)) for i in range(X.shape[1])])
```

```python
df = pd.concat([df, dfOneHot], axis=1)
df
```
